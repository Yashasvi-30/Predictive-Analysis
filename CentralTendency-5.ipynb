```python
#Q1	Measure of Central Tendency: Mean, Geometric mean, Harmonic mean, Mode, Median

import pandas as pd
import numpy as np
from scipy import stats

# Loading the dataset
data = pd.read_csv('bodyPerformance.csv')

# Selecting only numerical columns for calculating the mean
numerical_data = data.select_dtypes(include=np.number)

# Mean
mean = numerical_data.mean()

# Geometric Mean
geometric_mean = numerical_data.apply(lambda x: stats.gmean(x[x > 0]) if x[x > 0].size > 0 else np.nan)

# Harmonic Mean
harmonic_mean = numerical_data.apply(lambda x: stats.hmean(x[x > 0]) if x[x > 0].size > 0 else np.nan)

# Mode
mode = data.mode().iloc[0]

# Median
median = numerical_data.median()

mean, geometric_mean, harmonic_mean, mode, median
```

```python
#Q2 Measure of Dispersion: Variance, Standard deviation, Shape of Data (Symmetric, Skewness), Inter Quartile Range (IQR) / percentiles, Range, Mean Absolute Deviation(MAD)

# Selecting only numerical columns
numerical_data = data.select_dtypes(include=np.number)

# Variance and Standard Deviation
variance = numerical_data.var()
std_dev = numerical_data.std()

# Skewness
skewness = numerical_data.skew()

# IQR
IQR = numerical_data.quantile(0.75) - numerical_data.quantile(0.25)

# Range
data_range = numerical_data.max() - numerical_data.min()

# Mean Absolute Deviation
MAD = numerical_data.sub(numerical_data.mean()).abs().mean()

variance, std_dev, skewness, IQR, data_range, MAD
```

```python
#Q3	Correlation between features.

# Selecting only numerical columns for correlation analysis
numerical_data = data.select_dtypes(include=np.number)

# Correlation Matrix
correlation_matrix = numerical_data.corr()
correlation_matrix
```

```python
#Q4 Visualizing Data Distribution: Boxplot, Histograms, Density plots, Scatterplot, Bar chart

import matplotlib.pyplot as plt
import seaborn as sns

# Boxplot for each feature
plt.figure(figsize=(10, 5))
sns.boxplot(data=data)
plt.title("Boxplot of Each Feature")
plt.show()

# Histograms for each feature
data.hist(figsize=(10, 10), bins=20)
plt.suptitle("Histograms of Each Feature")
plt.show()

# Density Plots
# Calculating the number of numerical features for proper layout
num_numerical_features = len(data.select_dtypes(include=np.number).columns)

# Calculating rows and columns for subplot layout
num_rows = int(np.ceil(np.sqrt(num_numerical_features)))  # Calculating rows for layout
num_cols = int(np.ceil(num_numerical_features / num_rows))  # Calculating columns for layout

# Creating density plots with adjusted layout
data.plot(kind='density', subplots=True, layout=(num_rows, num_cols), sharex=False, figsize=(10, 10))  # Using calculated layout
plt.suptitle("Density Plots of Each Feature")
plt.show()

# Scatterplot Matrix
sns.pairplot(data)
plt.suptitle("Scatterplot Matrix")
plt.show()

# Bar chart (for a single categorical feature)
data['age'].value_counts().plot(kind='bar')
plt.title("Bar Chart of Age")
plt.show()
```
